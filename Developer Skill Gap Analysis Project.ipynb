```python
# Define the fill_mode function to fill NaNs with mode or 'Unknown' if mode is not found
def fill_mode(series):
    mode = series.mode()
    return series.fillna(mode[0]) if not mode.empty else series.fillna('Unknown')

# Copy the DataFrame to avoid modifying the original
filled_df = skill_gap_analysis.copy()

try:
    # Primary filling steps with specific groupings and ensuring categorical data type
    filled_df['PlatformWantToWorkWith'] = (
        filled_df.groupby(['PlatformHaveWorkedWith', 'DevType'])['PlatformWantToWorkWith']
        .apply(lambda group: fill_mode(group))
        .astype("category")  # Ensure data type remains categorical
    )

    filled_df['DatabaseWantToWorkWith'] = (
        filled_df.groupby(['PlatformWantToWorkWith', 'DatabaseHaveWorkedWith', 'DevType'])['DatabaseWantToWorkWith']
        .apply(lambda group: fill_mode(group))
        .astype("category")
    )

    filled_df['PlatformHaveWorkedWith'] = (
        filled_df.groupby(['PlatformWantToWorkWith', 'EdLevel', 'DevType'])['PlatformHaveWorkedWith']
        .apply(lambda group: fill_mode(group))
        .astype("category")
    )

    filled_df['ToolsTechWantToWorkWith'] = (
        filled_df.groupby(['ToolsTechHaveWorkedWith', 'DevType'])['ToolsTechWantToWorkWith']
        .apply(lambda group: fill_mode(group))
        .astype("category")
    )

    filled_df['DatabaseHaveWorkedWith'] = (
        filled_df.groupby(['DatabaseWantToWorkWith', 'DevType'])['DatabaseHaveWorkedWith']
        .apply(lambda group: fill_mode(group))
        .astype("category")
    )

    filled_df['DevType'] = (
        filled_df.groupby(['ToolsTechHaveWorkedWith', 'ToolsTechWantToWorkWith', 'PlatformWantToWorkWith',
                           'DatabaseHaveWorkedWith', 'LanguageHaveWorkedWith', 'LanguageWantToWorkWith', 'EdLevel'])['DevType']
        .apply(lambda group: fill_mode(group))
        .astype("category")
    )

    filled_df['ToolsTechHaveWorkedWith'] = (
        filled_df.groupby(['ToolsTechWantToWorkWith', 'DevType'])['ToolsTechHaveWorkedWith']
        .apply(lambda group: fill_mode(group))
        .astype("category")
    )

    filled_df['LanguageWantToWorkWith'] = (
        filled_df.groupby(['LanguageHaveWorkedWith', 'DevType'])['LanguageWantToWorkWith']
        .apply(lambda group: fill_mode(group))
        .astype("category")
    )

    filled_df['LanguageHaveWorkedWith'] = (
        filled_df.groupby(['LanguageWantToWorkWith', 'ToolsTechHaveWorkedWith', 'DevType'])['LanguageHaveWorkedWith']
        .apply(lambda group: fill_mode(group))
        .astype("category")
    )

    # Fallback fill for remaining nulls using broader groupings (e.g., only based on 'DevType')
    columns_to_fill = [
        'PlatformWantToWorkWith', 'DatabaseWantToWorkWith', 'PlatformHaveWorkedWith',
        'ToolsTechWantToWorkWith', 'DatabaseHaveWorkedWith', 'DevType',
        'ToolsTechHaveWorkedWith', 'LanguageWantToWorkWith', 'LanguageHaveWorkedWith'
    ]
    
    for col in columns_to_fill:
        filled_df[col] = filled_df.groupby('DevType')[col].apply(lambda group: fill_mode(group)).astype("category")

    # Final fill for remaining nulls by each individual column mode
    for col in columns_to_fill:
        filled_df[col] = fill_mode(filled_df[col]).astype("category")

    # Checking for null values after fallback filling
    print("\nNull values in each column after final filling:\n", filled_df.isnull().sum())

except Exception as e:
    print("An error occurred:", e)

# Verify data types to confirm columns remain categorical
print("\nData types after filling and ensuring categorical types:\n", filled_df.dtypes)
```

<h2>Gap Analysis by Country and Developer Type<h2>

```python
# Create columns for missing counts
Skill_gap_analysis['MissingLanguages'] = Skill_gap_analysis['LanguageGap'].apply(len)
Skill_gap_analysis['MissingDatabases'] = Skill_gap_analysis['DatabaseGap'].apply(len)

# Group by Country and DevType
gap_by_country_devtype = Skill_gap_analysis.groupby(['Country', 'DevType']).agg({
    'MissingLanguages': 'sum',
    'MissingDatabases': 'sum'
}).reset_index()

# Filter to top 5 countries
top_countries = (
    gap_by_country_devtype.groupby('Country')[['MissingLanguages', 'MissingDatabases']]
    .sum()
    .sort_values(by='MissingLanguages', ascending=False)
    .head(5)
    .index
)
filtered_data = gap_by_country_devtype[gap_by_country_devtype['Country'].isin(top_countries)]

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

countries = top_countries.tolist()
colors = sns.color_palette("tab10", len(countries) * 2)
color_map = {f"{country} - Languages": colors[i * 2] for i, country in enumerate(countries)}
color_map.update({f"{country} - Databases": colors[i * 2 + 1] for i, country in enumerate(countries)})

plt.figure(figsize=(14, 8))
for country in countries:
    subset = filtered_data[filtered_data['Country'] == country]
    plt.bar(
        subset['DevType'], subset['MissingLanguages'],
        label=f'{country} - Languages', color=color_map[f"{country} - Languages"], alpha=0.8
    )
    plt.bar(
        subset['DevType'], subset['MissingDatabases'],
        label=f'{country} - Databases', color=color_map[f"{country} - Databases"], alpha=0.6,
        bottom=subset['MissingLanguages']
    )
plt.title('Tech Talent Gap by Country and Developer Type (Top 5 Countries)', fontsize=14)
plt.xlabel('Developer Type', fontsize=12)
plt.ylabel('Number of Missing Skills', fontsize=12)
plt.legend(loc='upper left', fontsize=10, ncol=2)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Calculate the number of missing languages for each country from aggregated data
all_gaps['LanguageGapSize'] = all_gaps['LanguageGap'].apply(len)

# Filter out countries with no language gaps
filtered_gaps = all_gaps[all_gaps['LanguageGapSize'] > 0]

# Sort and select the top 10
top_10_language_gap = filtered_gaps.sort_values(by='LanguageGapSize', ascending=False).head(10)

# Plot the top 10 countries with the largest language gaps
plt.figure(figsize=(12, 8))  # Adjust the figure size for better visualization
sns.barplot(x='LanguageGapSize', y='Country', data=top_10_language_gap, palette='viridis')

# Add labels and formatting
plt.title('Top 10 Countries with the Largest Language Gaps', fontsize=16)
plt.xlabel('Number of Missing Languages', fontsize=14)
plt.ylabel('Country', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Combine all the lists of languages people want to work with
all_languages_wanted = sum(Skill_gap_analysis['LanguageWantToWorkWith'].dropna(), [])

# Count the frequency of each language
language_counts = Counter(all_languages_wanted)

# Get the top 10 most wanted languages
top_10_languages = language_counts.most_common(10)

# Separate the languages and their counts
languages, counts = zip(*top_10_languages)

# Plot the top 10 languages
plt.figure(figsize=(12, 8))
sns.barplot(x=list(counts), y=list(languages), palette='coolwarm')

# Add labels and formatting
plt.title('Top 10 Most Wanted Languages', fontsize=16)
plt.xlabel('Number of People Wanting to Work With', fontsize=14)
plt.ylabel('Language', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Combine all the lists of databases people want to work with
all_databases_wanted = sum(Skill_gap_analysis['DatabaseWantToWorkWith'].dropna(), [])

# Count the frequency of each database
database_counts = Counter(all_databases_wanted)

# Get the top 5 most wanted databases
top_5_databases = database_counts.most_common(5)

# Separate the databases and their counts
databases, db_counts = zip(*top_5_databases)

# Plot the top 5 databases
plt.figure(figsize=(10, 6))
sns.barplot(x=list(db_counts), y=list(databases), palette='Blues')

# Add labels and formatting
plt.title('Top 5 Most Wanted Databases', fontsize=16)
plt.xlabel('Number of People Wanting to Work With', fontsize=14)
plt.ylabel('Database', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

<h1>Key Insights and Trends from the Data on Tech Talent Skill Gap Analysis<h1>

Geographical Variation in Skill Gaps:

1) United States and India lead in missing skills, especially in programming languages and databases, indicating high demand for tech talent but a significant gap in skill supply.
    
2) Germany, Canada, and the UK also show noticeable gaps, though these are smaller compared to the U.S. and India. This suggests that while these countries are relatively developed in tech infrastructure, the rapidly evolving tech landscape still presents a challenge.
    
3) Countries like Yemen, Taiwan, and certain African nations show considerable gaps, potentially due to a combination of limited access to educational resources and growing demand for tech roles.
    
Role-Specific Skill Gaps:

1) Full-stack and backend developers exhibit the largest skill gaps, reflecting the complexity of these roles that require proficiency across multiple technologies.
    
2) Specialized roles such as data scientists, machine learning engineers, and DevOps specialists show smaller gaps but still indicate significant demand for skilled professionals in these areas.
    
3) Less Technical Roles, such as Academic Researchers or Marketing and Sales Professionals, show minimal skill gaps, suggesting that these positions do not heavily rely on technical skills, though there are still emerging needs for technical proficiency in related fields like digital marketing and sales analytics.
    
Programming Languages and Databases Dominate the Gaps:

1) The primary gap across countries and roles is in programming languages, followed by database management. Languages such as Python, JavaScript, Java, and SQL are particularly in high demand across various regions and roles.
    
Emerging Technologies Create Niche Skill Gaps:

1) Roles like mobile developers, game developers, and DevOps engineers show moderate skill gaps, likely due to the need for specialized knowledge in niche technologies like Unity, Swift, Kotlin, and cloud computing platforms.
    
Students Display Smaller Skill Gaps:

1) The skill gap for students is relatively smaller compared to experienced developers, reflecting their ongoing learning phase and exposure to a narrower set of technologies.
